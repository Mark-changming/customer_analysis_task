```python
import pandas as pd
import numpy as np
```

```python
df_order_products = pd.read_csv("./data/order_products/order_products.csv")
df_orders = pd.read_csv("./data/orders/orders.csv")
```

```python
df_orders
```

```python
df_order_products
```

```python
df_join = pd.merge(df_orders, df_order_products.drop_duplicates(subset=['order_id']), how="left", on="order_id")
df_join.head()
```

```python
grouped = df_join.groupby(["user_id"], as_index=True).order_number.max().reset_index()
grouped
```

```python
df_joinback = pd.merge(df_join, grouped, on="user_id", how="inner")
df_joinback
```

```python
df_joinback["eval_set"] = np.nan
df_joinback["eval_set"][df_joinback["product_id"].isnull()] = "test"
# joint["eval_set"][joint["product_id"].isnull()] = "test"
# df_joinback
```

```python
df_joinback.iloc[df_joinback[df_joinback["order_number_x"] == df_joinback["order_number_y"]].index, -1] = df_joinback.iloc[df_joinback[df_joinback["order_number_x"] == df_joinback["order_number_y"]].index, -1].fillna("train")
```

```python
df_joinback
```

```python
values = {'eval_set': "prior"}
df_joinback.fillna(value=values, inplace=True)
df_joinback
```

```python
train = df_joinback[df_joinback["eval_set"] == "train"]
train
```

```python
test = df_joinback[df_joinback["eval_set"] == "test"]
test
```
