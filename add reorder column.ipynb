```python
import numpy as np
import pandas as pd
```

```python
df_aisles = pd.read_csv("./data/aisles/aisles.csv")
df_departments = pd.read_csv("./data/departments/departments.csv")
df_order_products = pd.read_csv("./data/order_products/order_products.csv")
df_orders = pd.read_csv("./data/orders/orders.csv")
df_products = pd.read_csv("./data/products/products.csv")
```

```python
df_od = pd.merge(df_order_products, df_orders, on="order_id")
df_od.tail(10)
```

```python
grouped = df_od.groupby(["user_id", "product_id"], as_index=False).order_number.min()
grouped
```

```python
df_joinback = pd.merge(df_od, grouped, how="left", on=["user_id", "product_id"])
df_joinback
```

```python
df_joinback.loc[df_joinback["order_number_x"] == df_joinback["order_number_y"], "reorder"] = 1
df_joinback.loc[df_joinback["order_number_x"] != df_joinback["order_number_y"], "reorder"] = 0
df_joinback
```
